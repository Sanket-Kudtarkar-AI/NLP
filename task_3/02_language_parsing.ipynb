# Chunking Noun Phrases

```python
chunk_grammer_np = 'NP : {<DT>?<JJ>*<NN>}'
```

```python
chunk_parser_np = RegexpParser(chunk_grammer_np)
```

```python
np_chunked_oz = list()
for pos_tagged_sentence in pos_tagged_oz:
    np_chunked_oz.append(chunk_parser_np.parse(pos_tagged_sentence))
    
```

```python
chunks = list()

chunked_sentences = np_chunked_oz[:500]
for chunked_sentence in chunked_sentences:
        for subtree in chunked_sentence.subtrees(filter=lambda t: t.label() == 'NP'):
            chunks.append(tuple(subtree))
chunks
```

```python
from collections import Counter
chunk_counter = Counter()
for chunk in chunks:
    chunk_counter[chunk] += 1
```
